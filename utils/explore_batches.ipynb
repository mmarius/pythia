```python
import numpy as np
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/pythia-70m")
```

```python
# Note that according to the discussion here (https://github.com/EleutherAI/pythia/issues/123)
# the sequences come without EOD tokens
```

```python
start_iter = 0
end_iter = 1 # including
```

```python
path_to_batches = "/mnt/research/scratch/users/mmosba/pythia-data/batches"
input_ids_file = f"{path_to_batches}/input_ids-from{start_iter}-to{end_iter}.npy"
input_ids = np.load(input_ids_file)
print(input_ids.shape)
# These input_ids contain tokenized sequences of integers of size (None, 2049), where an integer corresponds to a unique token index. 
# Note that documents are concatenated and saperated by an EOD token. 
# Thus, each sample or batch may not start with an EOD token.
```
